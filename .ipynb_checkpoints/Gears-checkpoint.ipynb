# Gear Ratio

* When two gears of same diameter have one to one ratio. Which means everytime a driver gear completes full roation the driven gear also rotates full rotation.
* ### Driven size / Driver size = Gear ratio

* If driven gear is half the diameter of the driver gear then we have 1/2 ratio that means every time the driver gear completes a full rotation the driven gear completes 2 rotations.
* <img src="data/images/6.png" alt="Image description" style="width:500px;">
* And if its the opposite case where driver gear is half of the driven gear then the ration is 2 : 1 where driver gear makes 2 rotation anddriven makes only 1

# Compound Gear Train

* Let's designate the smaller gear attached as Gear A and the larger gear on the compound gear as Gear B. When the output gear (or driver gear) rotates at 150 RPM and produces 20 Nm of torque, due to a gear ratio of 1, Gear A will rotate at the same speed. Even though Gear B has larger teeth, since Gear B and Gear A share the same shaft, both will have the same RPM and torque.
